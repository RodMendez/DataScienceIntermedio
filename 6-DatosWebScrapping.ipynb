```python
# Finding elements
# Making a GET request
r = requests.get('https://www.geeksforgeeks.org/python-programming-language/')

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

# extrayendo contenido de las secciones div
s = soup.find('div', class_='entry-content')
# p se refiere a los parrafos
content = s.find_all('p')

print(content)
```

```python
# Finding by id
s = soup.find('div', id= 'main')

# Getting the leftbar
leftbar = s.find('ul', class_='leftBarList')

# All the li under the above ul
content = leftbar.find_all('li')

print(content)
```

```python
## quitando tags
s = soup.find('div', class_='entry-content')

lines = s.find_all('p')

for line in lines:
    print(line.text)
#     print(type(line))
#     print(line.text)
```

Buscando links

```python
# find all the anchor tags with "href"
for link in soup.find_all('a'):
    print(link.get('href'))
```

Imagenes

```python
images_list = []

images = soup.select('img')
for image in images:
    src = image.get('src')
    alt = image.get('alt')
    images_list.append({"src": src, "alt": alt})

for image in images_list:
    print(image)
```

## Ejemplo 2: Guardando csv

```python
all_quotes = []
for i in range(1, 11):
    url = f'https://quotes.toscrape.com/page/{i}/'
    page = urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')
    # de explorar el html de la página web, vemos que nos interesa el contenido de
    # las clases quote
    quotes = soup.find_all('div', class_='quote')

    for quote in quotes:
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        tags = quote.find('div', class_='tags').find_all('a')

        tags_list = []
        for tag in tags:
            tags_list.append(tag.text)

        single_quote = [text, author, tags_list]
        all_quotes.append(single_quote)

df = pd.DataFrame(all_quotes, columns=['quote', 'author', 'tags'])
df.to_csv('quotes.csv', index=False)
```

```python
while True: # Create a little chatbot
    keyword = input("\n\nHola, ¿Qué te gustaría buscar?")
    print(f"\n\nBuscando: {keyword}...\n\n")
    try:
        url = f"https://en.wikipedia.org/wiki/{keyword}"
        page = requests.get(url)
        text_info = BeautifulSoup(page.content, 'html.parser').find_all("p")
        print("Encontré esto: \n\n")
        for i in text_info:
            if (i.text != "\n"):
                print(i.text)
                break
        ans = input("\n\n¿Es suficiente? S (sí) N (no) ")
        if (ans.strip().lower() == "s"):
            print("\n\nMe da gusto. Ha sido un placer.")
            break
    except:
        print("Disculpa. No logré encontrar nada más.")
```
